```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.tools.plotting import andrews_curves
from pandas.tools.plotting import parallel_coordinates
from pandas.tools.plotting import radviz
```

```python
Schools = pd.read_csv("Public_Schools.csv")
Schools.head()
```

```python
Schools.info()
```

```python
sns.jointplot(x = "OBJECTID_1", y = "OBJECTID", data = Schools, size = 5)
```

```python
sns.FacetGrid(Schools, hue = "PL", size = 5)\
     .map(plt.scatter, "OBJECTID_1", "OBJECTID")\
     .add_legend()
```

```python
sns.boxplot(x = "PL", y = "OBJECTID_1", data = Schools)
```

```python
ax = sns.boxplot(x = "PL", y = "OBJECTID_1", data = Schools)
ax = sns.stripplot(x = "PL", y = "OBJECTID_1", data = Schools, jitter = True, edgecolor = "gray")
```

```python
sns.violinplot(x = "PL", y = "OBJECTID_1", data = Schools, size = 6)
```

```python
sns.FacetGrid(Schools, hue = "PL", size = 6)\
     .map(sns.kdeplot, "OBJECTID_1")\
     .add_legend()
```

```python
sns.pairplot(Schools.drop("OBJECTID",axis = 1), hue = "PL", size = 3)
```

```python
sns.pairplot(Schools.drop("OBJECTID", axis = 1), hue = "PL", size = 3, diag_kind = "kde")
```

```python
Schools.drop("OBJECTID", axis = 1).boxplot(by = "PL", figsize = (12, 6))
```
